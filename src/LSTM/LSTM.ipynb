```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(42)
tf.random.set_seed(42)

print("Tensorflow version: ", tf.__version__)
```

```python
# ====================================
# 1. DATA LOADING AND PREPROCESSING
# ====================================

def load_and_explore_data():
    print("Loading data...")
    
    # Load datasets
    train_df = pd.read_csv('train.csv')
    test_df = pd.read_csv('test.csv')
    valid_df = pd.read_csv('valid.csv')
    
    print("\n=== Data Overview ===")
    print(f"Train data shape: {train_df.shape}")
    print(f"Test data shape: {test_df.shape}")
    print(f"Valid data shape: {valid_df.shape}")
    
    print("\n=== Label Distribution ===")
    print("Train:")
    print(train_df['label'].value_counts())
    print("\nTest:")
    print(test_df['label'].value_counts())
    print("\nValid:")
    print(valid_df['label'].value_counts())
    
    return train_df, test_df, valid_df

def preprocess_data(train_df, test_df, valid_df, max_features=10000, max_length=100):
    
    all_texts = pd.concat([train_df['text'], test_df['text'], valid_df['text']]).values
    
    vectorizer = layers.TextVectorization(
        max_tokens=max_features,
        output_sequence_length=max_length,
        output_mode='int'
    )
    
    vectorizer.adapt(all_texts)
    
    X_train = vectorizer(train_df['text'].values)
    X_test = vectorizer(test_df['text'].values)
    X_valid = vectorizer(valid_df['text'].values)
    
    label_encoder = {'negative': 0, 'neutral': 1, 'positive': 2}
    y_train = np.array([label_encoder[label] for label in train_df['label'].values])
    y_test = np.array([label_encoder[label] for label in test_df['label'].values])
    y_valid = np.array([label_encoder[label] for label in valid_df['label'].values])
    
    print(f"Vocabulary size: {vectorizer.vocabulary_size()}")
    print(f"Sequence length: {max_length}")
    print(f"Number of classes: {len(label_encoder)}")
    
    return (X_train, y_train), (X_test, y_test), (X_valid, y_valid), vectorizer, label_encoder

# ====================================
# 2. LSTM MODEL BUILDING FUNCTIONS
# ====================================

def create_lstm_model(vocab_size, embedding_dim=128, lstm_units=64, num_lstm_layers=1, 
                      bidirectional=False, dropout_rate=0.3, num_classes=3, max_length=100):
    
    
    model = keras.Sequential()
    
    model.add(layers.Embedding(vocab_size, embedding_dim, input_length=max_length))
    
    for i in range(num_lstm_layers):
        return_sequences = (i < num_lstm_layers - 1)  
        
        if bidirectional:
            model.add(layers.Bidirectional(
                layers.LSTM(lstm_units, return_sequences=return_sequences, dropout=dropout_rate, 
                           recurrent_dropout=dropout_rate)
            ))
        else:
            model.add(layers.LSTM(lstm_units, return_sequences=return_sequences, dropout=dropout_rate,
                                 recurrent_dropout=dropout_rate))
        
        if i < num_lstm_layers - 1:
            model.add(layers.Dropout(dropout_rate))
    
    model.add(layers.Dropout(dropout_rate))
    
    model.add(layers.Dense(num_classes, activation='softmax'))
    
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_and_evaluate_model(model, X_train, y_train, X_valid, y_valid, X_test, y_test, 
                           epochs=20, batch_size=32, model_name="model"):
    
    print(f"\n=== Training {model_name} ===")
    print(model.summary())
    
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=5, restore_best_weights=True
    )
    
    reduce_lr = keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7
    )
    
    history = model.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_valid, y_valid),
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    f1_macro = f1_score(y_test, y_pred_classes, average='macro')
    
    print(f"\n=== Results for {model_name} ===")
    print(f"Macro F1-Score: {f1_macro:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_classes, 
                              target_names=['negative', 'neutral', 'positive']))
    
    return model, history, f1_macro, y_pred_classes

def plot_training_history(histories, labels, title):
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 2, 1)
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} - Train')
        plt.plot(history.history['val_loss'], label=f'{label} - Val', linestyle='--')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    for history, label in zip(histories, labels):
        plt.plot(history.history['accuracy'], label=f'{label} - Train')
        plt.plot(history.history['val_accuracy'], label=f'{label} - Val', linestyle='--')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ====================================
# 3. HYPERPARAMETER ANALYSIS
# ====================================

def analyze_num_layers(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    print("\n" + "="*50)
    print("ANALISIS PENGARUH JUMLAH LAYER LSTM")
    print("="*50)
    
    layer_configs = [1, 2, 3]
    models = []
    histories = []
    f1_scores = []
    
    for num_layers in layer_configs:
        model = create_lstm_model(
            vocab_size=vocab_size,
            num_lstm_layers=num_layers,
            lstm_units=64
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"LSTM_{num_layers}_layers"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [f"{n} layers" for n in layer_configs], 
                         "Number of LSTM Layers Comparison")
    
    plt.figure(figsize=(8, 6))
    plt.bar([f"{n} layers" for n in layer_configs], f1_scores, 
            color=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title('Macro F1-Score vs Number of LSTM Layers')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Kesimpulan
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN JUMLAH LAYER LSTM ===")
    print(f"Konfigurasi terbaik: {layer_configs[best_idx]} layer(s)")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Model dengan 1 layer LSTM memberikan performa terbaik")
        print("- Model yang lebih dalam mungkin mengalami overfitting")
    elif best_idx == 1:
        print("- Model dengan 2 layer LSTM memberikan performa terbaik")
        print("- Kedalaman moderate memberikan keseimbangan yang baik")
    else:
        print("- Model dengan 3 layer LSTM memberikan performa terbaik")
        print("- Model yang lebih dalam dapat menangkap pola yang lebih kompleks")
    
    return models[best_idx]

def analyze_lstm_units(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    """Analisis pengaruh banyak cell LSTM per layer"""
    print("\n" + "="*50)
    print("ANALISIS PENGARUH BANYAK CELL LSTM PER LAYER")
    print("="*50)
    
    unit_configs = [32, 64, 128]
    models = []
    histories = []
    f1_scores = []
    
    for units in unit_configs:
        model = create_lstm_model(
            vocab_size=vocab_size,
            lstm_units=units,
            num_lstm_layers=2
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"LSTM_{units}_units"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [f"{n} units" for n in unit_configs], 
                         "Number of LSTM Units Comparison")
    
    plt.figure(figsize=(8, 6))
    plt.bar([f"{n} units" for n in unit_configs], f1_scores, 
            color=['lightblue', 'lightgreen', 'lightyellow'])
    plt.title('Macro F1-Score vs Number of LSTM Units')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN BANYAK CELL LSTM PER LAYER ===")
    print(f"Konfigurasi terbaik: {unit_configs[best_idx]} units")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Model dengan 32 units memberikan performa terbaik")
        print("- Model yang lebih kecil cukup untuk dataset ini")
    elif best_idx == 1:
        print("- Model dengan 64 units memberikan performa terbaik")
        print("- Ukuran moderate memberikan keseimbangan yang baik")
    else:
        print("- Model dengan 128 units memberikan performa terbaik")
        print("- Model yang lebih besar dapat menangkap representasi yang lebih kaya")
    
    return models[best_idx]

def analyze_bidirectional(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    """Analisis pengaruh jenis layer LSTM berdasarkan arah"""
    print("\n" + "="*50)
    print("ANALISIS PENGARUH JENIS LAYER LSTM (ARAH)")
    print("="*50)
    
    configs = [
        {"bidirectional": False, "name": "Unidirectional"},
        {"bidirectional": True, "name": "Bidirectional"}
    ]
    
    models = []
    histories = []
    f1_scores = []
    
    for config in configs:
        model = create_lstm_model(
            vocab_size=vocab_size,
            bidirectional=config["bidirectional"],
            lstm_units=64,
            num_lstm_layers=2
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"LSTM_{config['name']}"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [config["name"] for config in configs], 
                         "Bidirectional vs Unidirectional LSTM Comparison")
    
    plt.figure(figsize=(8, 6))
    names = [config["name"] for config in configs]
    plt.bar(names, f1_scores, color=['lightcoral', 'lightblue'])
    plt.title('Macro F1-Score: Bidirectional vs Unidirectional LSTM')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN JENIS LAYER LSTM BERDASARKAN ARAH ===")
    print(f"Konfigurasi terbaik: {configs[best_idx]['name']}")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Unidirectional LSTM memberikan performa terbaik")
        print("- Model yang lebih sederhana sudah cukup untuk task ini")
        print("- Bidirectional mungkin menyebabkan overfitting")
    else:
        print("- Bidirectional LSTM memberikan performa terbaik")
        print("- Kemampuan melihat konteks masa depan membantu klasifikasi")
        print("- Informasi dari kedua arah memberikan representasi yang lebih kaya")
    
    return models[best_idx]

# ====================================
# 4. LSTM FORWARD PROPAGATION FROM SCRATCH
# ====================================

class LSTMFromScratch:
    """Implementasi forward propagation LSTM from scratch"""
    
    def __init__(self, model, vectorizer):
        self.model = model
        self.vectorizer = vectorizer
        self.weights = self.extract_weights()
        self.layer_info = self.get_layer_info()
        
    def extract_weights(self):
        """Extract weights dari trained Keras model"""
        weights = {}
        
        print("\n=== EXTRACTING MODEL WEIGHTS ===")
        
        for i, layer in enumerate(self.model.layers):
            layer_name = f"layer_{i}_{layer.__class__.__name__}"
            layer_type = layer.__class__.__name__
            
            if hasattr(layer, 'get_weights') and layer.get_weights():
                layer_weights = layer.get_weights()
                weights[layer_name] = layer_weights
                
                print(f"Layer {i} ({layer_type}):")
                for j, w in enumerate(layer_weights):
                    print(f"  Weight {j}: {w.shape}")
                    
        return weights
    
    def get_layer_info(self):
        """Get layer information"""
        layer_info = []
        
        for i, layer in enumerate(self.model.layers):
            layer_info.append({
                'index': i,
                'name': f"layer_{i}_{layer.__class__.__name__}",
                'type': layer.__class__.__name__,
                'layer': layer
            })
        
        return layer_info
    
    def sigmoid(self, x):
        """Sigmoid activation function"""
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))
    
    def tanh(self, x):
        """Tanh activation function"""
        return np.tanh(x)
    
    def softmax(self, x):
        """Softmax activation function dengan numerical stability"""
        if x.ndim == 1:
            x = x.reshape(1, -1)
        
        x_stable = x - np.max(x, axis=-1, keepdims=True)
        exp_x = np.exp(x_stable)
        return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
    
    def embedding_forward(self, input_ids, embedding_weights):
        """Forward pass untuk embedding layer"""
        if input_ids.ndim == 1:
            input_ids = input_ids.reshape(1, -1)
        
        return embedding_weights[input_ids]
    
    def lstm_cell_forward(self, x, h_prev, c_prev, weights):
        """Forward pass untuk single LSTM cell"""
        # LSTM gates computation
        # Keras menggunakan format: [W_xi, W_xf, W_xc, W_xo]
        # dan [W_hi, W_hf, W_hc, W_ho] untuk recurrent weights
        
        W_input, W_recurrent, b = weights[0], weights[1], weights[2]
        
        # Ensure proper shapes
        if x.ndim == 1:
            x = x.reshape(1, -1)
        if h_prev.ndim == 1:
            h_prev = h_prev.reshape(1, -1)
        if c_prev.ndim == 1:
            c_prev = c_prev.reshape(1, -1)
        
        # Split weights untuk 4 gates (input, forget, cell, output)
        units = W_recurrent.shape[0]
        
        W_xi = W_input[:, :units]
        W_hi = W_recurrent[:, :units]
        b_i = b[:units]
        
        W_xf = W_input[:, units:2*units]
        W_hf = W_recurrent[:, units:2*units]
        b_f = b[units:2*units]
        
        W_xc = W_input[:, 2*units:3*units]
        W_hc = W_recurrent[:, 2*units:3*units]
        b_c = b[2*units:3*units]
        
        W_xo = W_input[:, 3*units:]
        W_ho = W_recurrent[:, 3*units:]
        b_o = b[3*units:]
        
        i_t = self.sigmoid(np.dot(x, W_xi) + np.dot(h_prev, W_hi) + b_i)
        f_t = self.sigmoid(np.dot(x, W_xf) + np.dot(h_prev, W_hf) + b_f)
        c_tilde = self.tanh(np.dot(x, W_xc) + np.dot(h_prev, W_hc) + b_c)
        o_t = self.sigmoid(np.dot(x, W_xo) + np.dot(h_prev, W_ho) + b_o)
        
        c_t = f_t * c_prev + i_t * c_tilde
        h_t = o_t * self.tanh(c_t)
        
        return h_t, c_t
    
    def lstm_forward(self, x, weights, return_sequences=False):
        if x.ndim == 2:  # Sesuaikan batch dimension supaya sesuai format
            x = x.reshape(1, x.shape[0], x.shape[1])
            
        batch_size, seq_len, input_dim = x.shape
        
        hidden_dim = weights[1].shape[0] 
        
        h = np.zeros((batch_size, hidden_dim))
        c = np.zeros((batch_size, hidden_dim))
        
        if return_sequences:
            outputs = np.zeros((batch_size, seq_len, hidden_dim))
            for t in range(seq_len):
                h, c = self.lstm_cell_forward(x[:, t, :], h, c, weights)
                outputs[:, t, :] = h
            return outputs
        else:
            for t in range(seq_len):
                h, c = self.lstm_cell_forward(x[:, t, :], h, c, weights)
            return h
    
    def bidirectional_lstm_forward(self, x, weights):
        
        if x.ndim == 2:
            x = x.reshape(1, x.shape[0], x.shape[1])
            
        num_weights_per_direction = len(weights) // 2
        forward_weights = weights[:num_weights_per_direction]
        backward_weights = weights[num_weights_per_direction:]
        
        forward_out = self.lstm_forward(x, forward_weights, return_sequences=False)
        
        x_reversed = x[:, ::-1, :]
        backward_out = self.lstm_forward(x_reversed, backward_weights, return_sequences=False)
        
        if forward_out.ndim == 1:
            forward_out = forward_out.reshape(1, -1)
        if backward_out.ndim == 1:
            backward_out = backward_out.reshape(1, -1)
            
        combined_out = np.concatenate([forward_out, backward_out], axis=1)
        return combined_out
    
    def dense_forward(self, x, weights):
        W, b = weights[0], weights[1]
        
        if x.ndim == 1:
            x = x.reshape(1, -1)
            
        return np.dot(x, W) + b
    
    def predict(self, X_test):
        print(f"\nPredicting for {X_test.shape[0]} samples...")
        
        predictions = []
        
        for i in range(X_test.shape[0]):
            # Get single sample
            input_ids = X_test[i:i+1]  # Keep batch dimension, jangan X_test[i] pasti error karena dimensinya gak sesuai format
            current_output = input_ids
            
            # Forward pass semua layer
            for layer_info in self.layer_info:
                layer_name = layer_info['name']
                layer_type = layer_info['type']
                
                if layer_type == 'Embedding':
                    if layer_name in self.weights:
                        embedding_weights = self.weights[layer_name][0]
                        current_output = self.embedding_forward(current_output, embedding_weights)
                
                elif layer_type == 'LSTM':
                    if layer_name in self.weights:
                        is_last_lstm = True
                        for j in range(layer_info['index']+1, len(self.layer_info)):
                            if self.layer_info[j]['type'] in ['LSTM', 'Bidirectional']:
                                is_last_lstm = False
                                break
                        
                        return_sequences = not is_last_lstm
                        current_output = self.lstm_forward(
                            current_output, 
                            self.weights[layer_name], 
                            return_sequences=return_sequences
                        )
                
                elif layer_type == 'Bidirectional':
                    if layer_name in self.weights:
                        current_output = self.bidirectional_lstm_forward(
                            current_output, 
                            self.weights[layer_name]
                        )
                
                elif layer_type == 'Dropout':
                    # Skip dropout in inference
                    continue
                
                elif layer_type == 'Dense':
                    if layer_name in self.weights:
                        current_output = self.dense_forward(current_output, self.weights[layer_name])
                        
                        # Apply softmax if this is the last layer (output layer)
                        if layer_info['index'] == len(self.layer_info) - 1:
                            current_output = self.softmax(current_output)
            
            predictions.append(current_output[0])  # Remove batch dimension
            
            if (i + 1) % 100 == 0:
                print(f"  Processed {i + 1}/{X_test.shape[0]} samples")
        
        return np.array(predictions)

def compare_implementations(keras_model, vectorizer, X_test, y_test):
    """Perbandingan antara Keras vs From Scratch dengan data test yang sama"""
    print("\n" + "="*60)
    print("PERBANDINGAN KERAS VS FROM SCRATCH - LSTM")
    print("="*60)

    keras_pred = keras_model.predict(X_test, verbose=0)
    keras_pred_classes = np.argmax(keras_pred, axis=1)
    
    print("Creating LSTM from scratch implementation...")
    lstm_scratch = LSTMFromScratch(keras_model, vectorizer)
    

    scratch_pred = lstm_scratch.predict(X_test)
    scratch_pred_classes = np.argmax(scratch_pred, axis=1)
    
    keras_f1 = f1_score(y_test, keras_pred_classes, average='macro')
    scratch_f1 = f1_score(y_test, scratch_pred_classes, average='macro')
    
    print(f"\n=== COMPARISON RESULTS ===")
    print(f"Keras F1-Score: {keras_f1:.4f}")
    print(f"From Scratch F1-Score: {scratch_f1:.4f}")
    print(f"Difference: {abs(keras_f1 - scratch_f1):.6f}")
    
    print(f"\n=== DETAILED COMPARISON ===")
    print(f"Keras predictions shape: {keras_pred.shape}")
    print(f"From scratch predictions shape: {scratch_pred.shape}")
    
    print(f"\nFirst 10 predictions comparison:")
    print("Sample | Keras Class | Scratch Class | Match | True Label")
    print("-" * 55)
    for i in range(10):
        match = "✓" if keras_pred_classes[i] == scratch_pred_classes[i] else "✗"
        print(f"{i+1:6d} | {keras_pred_classes[i]:11d} | {scratch_pred_classes[i]:13d} | {match:5s} | {y_test[i]:10d}")
    
    matches = np.sum(keras_pred_classes == scratch_pred_classes)
    accuracy = matches / len(keras_pred_classes)
    print(f"\nImplementation accuracy: {accuracy:.4f} ({matches}/{len(keras_pred_classes)} matches)")
    
    print(f"\n=== PREDICTION PROBABILITIES COMPARISON (First 5 samples) ===")
    for i in range(5):
        print(f"Sample {i+1}:")
        print(f"  Keras:       [{keras_pred[i][0]:.4f}, {keras_pred[i][1]:.4f}, {keras_pred[i][2]:.4f}]")
        print(f"  From Scratch: [{scratch_pred[i][0]:.4f}, {scratch_pred[i][1]:.4f}, {scratch_pred[i][2]:.4f}]")
        print(f"  Difference:   [{abs(keras_pred[i][0]-scratch_pred[i][0]):.4f}, {abs(keras_pred[i][1]-scratch_pred[i][1]):.4f}, {abs(keras_pred[i][2]-scratch_pred[i][2]):.4f}]")
        print()
    
    prob_diff = np.abs(keras_pred - scratch_pred)
    print(f"=== PROBABILITY DIFFERENCES STATISTICS ===")
    print(f"Mean absolute difference: {np.mean(prob_diff):.6f}")
    print(f"Max absolute difference: {np.max(prob_diff):.6f}")
    print(f"Std absolute difference: {np.std(prob_diff):.6f}")
    
    return keras_pred, scratch_pred, accuracy

# ====================================
# 5. VISUALIZATION AND ANALYSIS FUNCTIONS
# ====================================

def plot_confusion_matrices(models, predictions, y_test, model_names):
    n_models = len(models)
    fig, axes = plt.subplots(1, n_models, figsize=(6*n_models, 5))
    if n_models == 1:
        axes = [axes]
    
    class_names = ['Negative', 'Neutral', 'Positive']
    
    for i, (pred, name) in enumerate(zip(predictions, model_names)):
        pred_classes = np.argmax(pred, axis=1) if pred.ndim > 1 else pred
        cm = confusion_matrix(y_test, pred_classes)
        
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=class_names, yticklabels=class_names, ax=axes[i])
        axes[i].set_title(f'Confusion Matrix - {name}')
        axes[i].set_xlabel('Predicted')
        axes[i].set_ylabel('Actual')
    
    plt.tight_layout()
    plt.show()

def analyze_model_complexity(models, model_names):
    
    complexities = []
    for model, name in zip(models, model_names):
        total_params = model.count_params()
        trainable_params = sum([np.prod(w.shape) for w in model.trainable_weights])
        
        complexities.append({
            'name': name,
            'total_params': total_params,
            'trainable_params': trainable_params
        })
        
        print(f"\n{name}:")
        print(f"  Total parameters: {total_params:,}")
        print(f"  Trainable parameters: {trainable_params:,}")
    
    plt.figure(figsize=(10, 6))
    names = [c['name'] for c in complexities]
    params = [c['total_params'] for c in complexities]
    
    plt.bar(names, params, color=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow'][:len(names)])
    plt.title('Model Complexity Comparison (Total Parameters)')
    plt.ylabel('Number of Parameters')
    plt.xticks(rotation=45)
    
    for i, param in enumerate(params):
        plt.text(i, param + max(params)*0.01, f'{param:,}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    return complexities

def analyze_prediction_confidence(predictions, y_test, model_names):
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.flatten()
    
    for i, (pred, name) in enumerate(zip(predictions, model_names)):
        if i >= 4:
            break
            
        if pred.ndim > 1:
            confidences = np.max(pred, axis=1)
            pred_classes = np.argmax(pred, axis=1)
        else:
            confidences = np.ones_like(pred)
            pred_classes = pred
        
        correct_mask = pred_classes == y_test
        correct_conf = confidences[correct_mask]
        incorrect_conf = confidences[~correct_mask]
        
        axes[i].hist(correct_conf, bins=20, alpha=0.7, label='Correct', color='green')
        axes[i].hist(incorrect_conf, bins=20, alpha=0.7, label='Incorrect', color='red')
        axes[i].set_title(f'{name} - Prediction Confidence')
        axes[i].set_xlabel('Confidence (Max Probability)')
        axes[i].set_ylabel('Frequency')
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)
        
        print(f"\n{name}:")
        print(f"  Correct predictions - Mean confidence: {np.mean(correct_conf):.4f}")
        print(f"  Incorrect predictions - Mean confidence: {np.mean(incorrect_conf):.4f}")
        print(f"  Total accuracy: {len(correct_conf)/(len(correct_conf)+len(incorrect_conf)):.4f}")
    
    plt.tight_layout()
    plt.show()

# ====================================
# 6. MAIN EXECUTION FUNCTION
# ====================================

def main():
    print("="*80)
    print("LSTM NEURAL NETWORK")
    print("="*80)
    
    # Load and preprocess data
    train_df, test_df, valid_df = load_and_explore_data()
    (X_train, y_train), (X_test, y_test), (X_valid, y_valid), vectorizer, label_encoder = preprocess_data(
        train_df, test_df, valid_df
    )
    
    vocab_size = vectorizer.vocabulary_size()
    print(f"\nVocabulary size: {vocab_size}")
    
    all_models = []
    all_predictions = []
    all_model_names = []
    all_f1_scores = []
    
    # 1. Analyze number of LSTM layers
    print("\n" + "="*80)
    print("PHASE 1: ANALYZING NUMBER OF LSTM LAYERS")
    print("="*80)
    best_layers_model = analyze_num_layers(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    # 2. Analyze number of LSTM units
    print("\n" + "="*80)
    print("PHASE 2: ANALYZING NUMBER OF LSTM UNITS")
    print("="*80)
    best_units_model = analyze_lstm_units(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    # bidirectional vs unidirectional
    print("\n" + "="*80)
    print("PHASE 3: ANALYZING BIDIRECTIONAL VS UNIDIRECTIONAL")
    print("="*80)
    best_direction_model = analyze_bidirectional(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    print("\n" + "="*80)
    print("PHASE 4: CREATING FINAL OPTIMIZED MODEL")
    print("="*80)
    
    final_model = create_lstm_model(
        vocab_size=vocab_size,
        embedding_dim=128,
        lstm_units=64,
        num_lstm_layers=2,
        bidirectional=True,
        dropout_rate=0.3,
        num_classes=3
    )
    
    final_model, final_history, final_f1, final_pred_classes = train_and_evaluate_model(
        final_model, X_train, y_train, X_valid, y_valid, X_test, y_test,
        epochs=25, model_name="Final_Optimized_LSTM"
    )
    
    final_pred = final_model.predict(X_test, verbose=0)
    all_models.extend([best_layers_model, best_units_model, best_direction_model, final_model])
    all_predictions.extend([
        best_layers_model.predict(X_test, verbose=0),
        best_units_model.predict(X_test, verbose=0), 
        best_direction_model.predict(X_test, verbose=0),
        final_pred
    ])
    all_model_names.extend(["Best Layers", "Best Units", "Best Direction", "Final Optimized"])
    
    # Calculate F1 scores for all models
    for pred in all_predictions:
        pred_classes = np.argmax(pred, axis=1)
        f1 = f1_score(y_test, pred_classes, average='macro')
        all_f1_scores.append(f1)
    
    # Keras vs Scratch 
    print("\n" + "="*80)
    print("PHASE 5: COMPARING KERAS VS FROM SCRATCH IMPLEMENTATION")
    print("="*80)
    
    keras_pred, scratch_pred, implementation_accuracy = compare_implementations(
        final_model, vectorizer, X_test, y_test
    )
    #

    
    print("\n" + "="*80)
    print("PHASE 6: COMPREHENSIVE ANALYSIS AND VISUALIZATION")
    print("="*80)
    
    plot_confusion_matrices(all_models, all_predictions, y_test, all_model_names)
    
    complexities = analyze_model_complexity(all_models, all_model_names)
    
    analyze_prediction_confidence(all_predictions, y_test, all_model_names)
    
    print("\n" + "="*80)
    print("FINAL SUMMARY")
    print("="*80)
    
    best_model_idx = np.argmax(all_f1_scores)
    best_model_name = all_model_names[best_model_idx]
    best_f1_score = all_f1_scores[best_model_idx]
    
    print(f"\n=== BEST MODEL PERFORMANCE ===")
    print(f"Best Model: {best_model_name}")
    print(f"Best F1-Score: {best_f1_score:.4f}")
    
    print(f"\n=== ALL MODEL COMPARISON ===")
    for name, f1, complexity in zip(all_model_names, all_f1_scores, complexities):
        print(f"{name:20s}: F1={f1:.4f}, Params={complexity['total_params']:,}")
    
    print(f"\n=== IMPLEMENTATION VERIFICATION ===")
    print(f"From Scratch Implementation Accuracy: {implementation_accuracy:.4f}")
    if implementation_accuracy > 0.95:
        print("sangat akurat")
    elif implementation_accuracy > 0.90:
        print("Cukup akurat")
    else:
        print("cukup bagus")
    
    return {
        'models': all_models,
        'predictions': all_predictions,
        'model_names': all_model_names,
        'f1_scores': all_f1_scores,
        'best_model': all_models[best_model_idx],
        'best_model_name': best_model_name,
        'implementation_accuracy': implementation_accuracy,
        'vectorizer': vectorizer,
        'label_encoder': label_encoder
    }

# ====================================
# 7. ADDITIONAL UTILITY FUNCTIONS
# ====================================

def save_model_and_results(results, save_path="lstm_results"):
    """Save model dan hasil analisis"""
    import os
    os.makedirs(save_path, exist_ok=True)
    
    # Save best model
    results['best_model'].save(f"{save_path}/best_lstm_model.h5")
    
    # Save vectorizer
    with open(f"{save_path}/vectorizer.pkl", 'wb') as f:
        pickle.dump(results['vectorizer'], f)
    
    # Save label encoder
    with open(f"{save_path}/label_encoder.pkl", 'wb') as f:
        pickle.dump(results['label_encoder'], f)
    
    # Save results summary
    summary = {
        'model_names': results['model_names'],
        'f1_scores': results['f1_scores'],
        'best_model_name': results['best_model_name'],
        'implementation_accuracy': results['implementation_accuracy']
    }
    
    with open(f"{save_path}/results_summary.pkl", 'wb') as f:
        pickle.dump(summary, f)
    
    print(f"\nModel and results saved to {save_path}/")

# ====================================
# 8. EXECUTION
# ====================================

if __name__ == "__main__":
    # Run main analysis
    results = main()
    
    # Save results
    save_model_and_results(results)
```
