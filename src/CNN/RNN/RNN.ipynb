```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)

# ====================================
# 1. DATA LOADING AND PREPROCESSING
# ====================================

def load_and_explore_data():
    """Load dan eksplorasi data"""
    print("Loading data...")
    
    train_df = pd.read_csv('train.csv')
    test_df = pd.read_csv('test.csv')
    valid_df = pd.read_csv('valid.csv')
    
    print("\n=== Data Overview ===")
    print(f"Train data shape: {train_df.shape}")
    print(f"Test data shape: {test_df.shape}")
    print(f"Valid data shape: {valid_df.shape}")
    
    print("\n=== Label Distribution ===")
    print("Train:")
    print(train_df['label'].value_counts())
    print("\nTest:")
    print(test_df['label'].value_counts())
    print("\nValid:")
    print(valid_df['label'].value_counts())
    
    return train_df, test_df, valid_df

def preprocess_data(train_df, test_df, valid_df, max_features=10000, max_length=100):
    """Preprocessing data dengan TextVectorization"""
    print("\n=== Data Preprocessing ===")
    
    all_texts = pd.concat([train_df['text'], test_df['text'], valid_df['text']]).values
    
    vectorizer = layers.TextVectorization(
        max_tokens=max_features,
        output_sequence_length=max_length,
        output_mode='int'
    )
    
    vectorizer.adapt(all_texts)
    
    X_train = vectorizer(train_df['text'].values)
    X_test = vectorizer(test_df['text'].values)
    X_valid = vectorizer(valid_df['text'].values)
    
    label_encoder = {'negative': 0, 'neutral': 1, 'positive': 2}
    y_train = np.array([label_encoder[label] for label in train_df['label'].values])
    y_test = np.array([label_encoder[label] for label in test_df['label'].values])
    y_valid = np.array([label_encoder[label] for label in valid_df['label'].values])
    
    print(f"Vocabulary size: {vectorizer.vocabulary_size()}")
    print(f"Sequence length: {max_length}")
    print(f"Number of classes: {len(label_encoder)}")
    
    return (X_train, y_train), (X_test, y_test), (X_valid, y_valid), vectorizer, label_encoder

# ====================================
# 2. MODEL BUILDING FUNCTIONS
# ====================================

def create_rnn_model(vocab_size, embedding_dim=128, rnn_units=64, num_rnn_layers=1, 
                     bidirectional=False, dropout_rate=0.3, num_classes=3, max_length=100):
    """Create RNN model dengan berbagai konfigurasi"""
    
    model = keras.Sequential()
    
    # Embedding layer
    model.add(layers.Embedding(vocab_size, embedding_dim, input_length=max_length))
    
    # RNN layers
    for i in range(num_rnn_layers):
        return_sequences = (i < num_rnn_layers - 1)
        
        if bidirectional:
            model.add(layers.Bidirectional(
                layers.SimpleRNN(rnn_units, return_sequences=return_sequences, dropout=dropout_rate)
            ))
        else:
            model.add(layers.SimpleRNN(rnn_units, return_sequences=return_sequences, dropout=dropout_rate))
        
        if i < num_rnn_layers - 1:
            model.add(layers.Dropout(dropout_rate))
    
    model.add(layers.Dropout(dropout_rate))
    
    model.add(layers.Dense(num_classes, activation='softmax'))
    
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_and_evaluate_model(model, X_train, y_train, X_valid, y_valid, X_test, y_test, 
                           epochs=20, batch_size=32, model_name="model"):
    """Train dan evaluasi model"""
    
    print(f"\n=== Training {model_name} ===")
    print(model.summary())
    
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=5, restore_best_weights=True
    )
    
    reduce_lr = keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7
    )
    
    history = model.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(X_valid, y_valid),
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    f1_macro = f1_score(y_test, y_pred_classes, average='macro')
    
    print(f"\n=== Results for {model_name} ===")
    print(f"Macro F1-Score: {f1_macro:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_classes, 
                              target_names=['negative', 'neutral', 'positive']))
    
    return model, history, f1_macro, y_pred_classes

def plot_training_history(histories, labels, title):
    """Plot training dan validation loss"""
    plt.figure(figsize=(15, 5))
    
    # Training Loss
    plt.subplot(1, 2, 1)
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} - Train')
        plt.plot(history.history['val_loss'], label=f'{label} - Val', linestyle='--')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    for history, label in zip(histories, labels):
        plt.plot(history.history['accuracy'], label=f'{label} - Train')
        plt.plot(history.history['val_accuracy'], label=f'{label} - Val', linestyle='--')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ====================================
# 3. HYPERPARAMETER ANALYSIS
# ====================================

def analyze_num_layers(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    """Analisis pengaruh jumlah layer RNN"""
    print("\n" + "="*50)
    print("ANALISIS PENGARUH JUMLAH LAYER RNN")
    print("="*50)
    
    layer_configs = [1, 2, 3]
    models = []
    histories = []
    f1_scores = []
    
    for num_layers in layer_configs:
        model = create_rnn_model(
            vocab_size=vocab_size,
            num_rnn_layers=num_layers,
            rnn_units=64
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"RNN_{num_layers}_layers"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [f"{n} layers" for n in layer_configs], 
                         "Number of RNN Layers Comparison")
    
    plt.figure(figsize=(8, 6))
    plt.bar([f"{n} layers" for n in layer_configs], f1_scores, 
            color=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title('Macro F1-Score vs Number of RNN Layers')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN JUMLAH LAYER RNN ===")
    print(f"Konfigurasi terbaik: {layer_configs[best_idx]} layer(s)")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Model dengan 1 layer RNN memberikan performa terbaik")
        print("- Model yang lebih dalam mungkin mengalami overfitting")
    elif best_idx == 1:
        print("- Model dengan 2 layer RNN memberikan performa terbaik")
        print("- Kedalaman moderate memberikan keseimbangan yang baik")
    else:
        print("- Model dengan 3 layer RNN memberikan performa terbaik")
        print("- Model yang lebih dalam dapat menangkap pola yang lebih kompleks")
    
    return models[best_idx]

def analyze_rnn_units(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    """Analisis pengaruh banyak cell RNN per layer"""
    print("\n" + "="*50)
    print("ANALISIS PENGARUH BANYAK CELL RNN PER LAYER")
    print("="*50)
    
    unit_configs = [32, 64, 128]
    models = []
    histories = []
    f1_scores = []
    
    for units in unit_configs:
        model = create_rnn_model(
            vocab_size=vocab_size,
            rnn_units=units,
            num_rnn_layers=2
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"RNN_{units}_units"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [f"{n} units" for n in unit_configs], 
                         "Number of RNN Units Comparison")
    
    plt.figure(figsize=(8, 6))
    plt.bar([f"{n} units" for n in unit_configs], f1_scores, 
            color=['lightblue', 'lightgreen', 'lightyellow'])
    plt.title('Macro F1-Score vs Number of RNN Units')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN BANYAK CELL RNN PER LAYER ===")
    print(f"Konfigurasi terbaik: {unit_configs[best_idx]} units")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Model dengan 32 units memberikan performa terbaik")
        print("- Model yang lebih kecil cukup untuk dataset ini")
    elif best_idx == 1:
        print("- Model dengan 64 units memberikan performa terbaik")
        print("- Ukuran moderate memberikan keseimbangan yang baik")
    else:
        print("- Model dengan 128 units memberikan performa terbaik")
        print("- Model yang lebih besar dapat menangkap representasi yang lebih kaya")
    
    return models[best_idx]

def analyze_bidirectional(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size):
    """Analisis pengaruh jenis layer RNN berdasarkan arah"""
    print("\n" + "="*50)
    print("ANALISIS PENGARUH JENIS LAYER RNN (ARAH)")
    print("="*50)
    
    configs = [
        {"bidirectional": False, "name": "Unidirectional"},
        {"bidirectional": True, "name": "Bidirectional"}
    ]
    
    models = []
    histories = []
    f1_scores = []
    
    for config in configs:
        model = create_rnn_model(
            vocab_size=vocab_size,
            bidirectional=config["bidirectional"],
            rnn_units=64,
            num_rnn_layers=2
        )
        
        model, history, f1_score_val, _ = train_and_evaluate_model(
            model, X_train, y_train, X_valid, y_valid, X_test, y_test,
            model_name=f"RNN_{config['name']}"
        )
        
        models.append(model)
        histories.append(history)
        f1_scores.append(f1_score_val)
    
    plot_training_history(histories, [config["name"] for config in configs], 
                         "Bidirectional vs Unidirectional RNN Comparison")
    
    plt.figure(figsize=(8, 6))
    names = [config["name"] for config in configs]
    plt.bar(names, f1_scores, color=['lightcoral', 'lightblue'])
    plt.title('Macro F1-Score: Bidirectional vs Unidirectional RNN')
    plt.ylabel('Macro F1-Score')
    plt.ylim(0, 1)
    for i, score in enumerate(f1_scores):
        plt.text(i, score + 0.01, f'{score:.4f}', ha='center')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Kesimpulan
    best_idx = np.argmax(f1_scores)
    print(f"\n=== KESIMPULAN JENIS LAYER RNN BERDASARKAN ARAH ===")
    print(f"Konfigurasi terbaik: {configs[best_idx]['name']}")
    print(f"F1-Score terbaik: {f1_scores[best_idx]:.4f}")
    print("Analisis:")
    if best_idx == 0:
        print("- Unidirectional RNN memberikan performa terbaik")
        print("- Model yang lebih sederhana sudah cukup untuk task ini")
        print("- Bidirectional mungkin menyebabkan overfitting")
    else:
        print("- Bidirectional RNN memberikan performa terbaik")
        print("- Kemampuan melihat konteks masa depan membantu klasifikasi")
        print("- Informasi dari kedua arah memberikan representasi yang lebih banyak")
    
    return models[best_idx]

# ====================================
# 4. FORWARD PROPAGATION FROM SCRATCH
# ====================================

class SimpleRNNFromScratch:
    """Implementasi forward propagation Simple RNN from scratch"""
    
    def __init__(self, model, vectorizer):
        self.model = model
        self.vectorizer = vectorizer
        self.weights = self.extract_weights()
        self.layer_info = self.get_layer_info()
        
    def extract_weights(self):
        """Extract weights dari trained Keras model"""
        weights = {}
        
        print("\n=== EXTRACTING MODEL WEIGHTS ===")
        
        for i, layer in enumerate(self.model.layers):
            layer_name = f"layer_{i}_{layer.__class__.__name__}"
            layer_type = layer.__class__.__name__
            
            if hasattr(layer, 'get_weights') and layer.get_weights():
                layer_weights = layer.get_weights()
                weights[layer_name] = layer_weights
                
                print(f"Layer {i} ({layer_type}):")
                for j, w in enumerate(layer_weights):
                    print(f"  Weight {j}: {w.shape}")
                    
        return weights
    
    def get_layer_info(self):
        """Get layer information"""
        layer_info = []
        
        for i, layer in enumerate(self.model.layers):
            layer_info.append({
                'index': i,
                'name': f"layer_{i}_{layer.__class__.__name__}",
                'type': layer.__class__.__name__,
                'layer': layer
            })
        
        return layer_info
    
    def tanh(self, x):
        """Tanh activation function"""
        return np.tanh(x)
    
    def softmax(self, x):
        """Softmax activation function dengan numerical stability"""
        if x.ndim == 1:
            x = x.reshape(1, -1)
        
        # Subtract max for numerical stability
        x_stable = x - np.max(x, axis=-1, keepdims=True)
        exp_x = np.exp(x_stable)
        return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
    
    def embedding_forward(self, input_ids, embedding_weights):
        """Forward pass untuk embedding layer"""
        if input_ids.ndim == 1:
            input_ids = input_ids.reshape(1, -1)
        
        return embedding_weights[input_ids]
    
    def simple_rnn_cell_forward(self, x, h_prev, weights):
        """Forward pass untuk single Simple RNN cell"""
        # Simple RNN: h_t = tanh(W_x * x_t + W_h * h_{t-1} + b)
        
        W_x, W_h, b = weights[0], weights[1], weights[2]
        
        # Ensure proper shapes
        if x.ndim == 1:
            x = x.reshape(1, -1)
        if h_prev.ndim == 1:
            h_prev = h_prev.reshape(1, -1)
        
        # Simple RNN computation
        h_t = self.tanh(np.dot(x, W_x) + np.dot(h_prev, W_h) + b)
        
        return h_t
    
    def simple_rnn_forward(self, x, weights, return_sequences=False):
        """Forward pass untuk Simple RNN layer"""
        if x.ndim == 2:  # Add batch dimension if needed
            x = x.reshape(1, x.shape[0], x.shape[1])
            
        batch_size, seq_len, input_dim = x.shape
        
        # Get hidden dimension from weights
        hidden_dim = weights[1].shape[0]  # From recurrent weights W_h
        
        # Initialize hidden state
        h = np.zeros((batch_size, hidden_dim))
        
        if return_sequences:
            outputs = np.zeros((batch_size, seq_len, hidden_dim))
            for t in range(seq_len):
                h = self.simple_rnn_cell_forward(x[:, t, :], h, weights)
                outputs[:, t, :] = h
            return outputs
        else:
            for t in range(seq_len):
                h = self.simple_rnn_cell_forward(x[:, t, :], h, weights)
            return h
    
    def bidirectional_simple_rnn_forward(self, x, weights):
        """Forward pass untuk Bidirectional Simple RNN layer"""
        # Ensure x is 3D
        if x.ndim == 2:
            x = x.reshape(1, x.shape[0], x.shape[1])
            
        # Split weights untuk forward dan backward
        num_weights_per_direction = len(weights) // 2
        forward_weights = weights[:num_weights_per_direction]
        backward_weights = weights[num_weights_per_direction:]
        
        # Forward direction
        forward_out = self.simple_rnn_forward(x, forward_weights, return_sequences=False)
        
        # Backward direction (reverse sequence)
        x_reversed = x[:, ::-1, :]  # Now x is guaranteed to be 3D
        backward_out = self.simple_rnn_forward(x_reversed, backward_weights, return_sequences=False)
        
        # Concatenate outputs
        if forward_out.ndim == 1:
            forward_out = forward_out.reshape(1, -1)
        if backward_out.ndim == 1:
            backward_out = backward_out.reshape(1, -1)
            
        combined_out = np.concatenate([forward_out, backward_out], axis=1)
        return combined_out
    
    def dense_forward(self, x, weights):
        """Forward pass untuk dense layer"""
        W, b = weights[0], weights[1]
        
        if x.ndim == 1:
            x = x.reshape(1, -1)
            
        return np.dot(x, W) + b
    
    def predict(self, X_test):
        """Predict untuk test data menggunakan vectorized input yang sama dengan Keras"""
        print(f"\nPredicting for {X_test.shape[0]} samples...")
        
        predictions = []
        
        for i in range(X_test.shape[0]):
            # Get single sample
            input_ids = X_test[i:i+1]  # Keep batch dimension
            current_output = input_ids
            
            # Forward pass through each layer
            for layer_info in self.layer_info:
                layer_name = layer_info['name']
                layer_type = layer_info['type']
                
                if layer_type == 'Embedding':
                    if layer_name in self.weights:
                        embedding_weights = self.weights[layer_name][0]
                        current_output = self.embedding_forward(current_output, embedding_weights)
                
                elif layer_type == 'SimpleRNN':
                    if layer_name in self.weights:
                        # Check if this is the last RNN layer
                        is_last_rnn = True
                        for j in range(layer_info['index']+1, len(self.layer_info)):
                            if self.layer_info[j]['type'] in ['SimpleRNN', 'Bidirectional']:
                                is_last_rnn = False
                                break
                        
                        return_sequences = not is_last_rnn
                        current_output = self.simple_rnn_forward(
                            current_output, 
                            self.weights[layer_name], 
                            return_sequences=return_sequences
                        )
                
                elif layer_type == 'Bidirectional':
                    if layer_name in self.weights:
                        current_output = self.bidirectional_simple_rnn_forward(
                            current_output, 
                            self.weights[layer_name]
                        )
                
                elif layer_type == 'Dropout':
                    # Skip dropout in inference
                    continue
                
                elif layer_type == 'Dense':
                    if layer_name in self.weights:
                        current_output = self.dense_forward(current_output, self.weights[layer_name])
                        
                        # Apply softmax if this is the last layer (output layer)
                        if layer_info['index'] == len(self.layer_info) - 1:
                            current_output = self.softmax(current_output)
            
            predictions.append(current_output[0])  # Remove batch dimension
            
            if (i + 1) % 100 == 0:
                print(f"  Processed {i + 1}/{X_test.shape[0]} samples")
        
        return np.array(predictions)

def compare_implementations(keras_model, vectorizer, X_test, y_test):
    """Perbandingan antara Keras vs From Scratch dengan data test yang sama"""
    print("\n" + "="*60)
    print("PERBANDINGAN KERAS VS FROM SCRATCH - SIMPLE RNN")
    print("="*60)
    
    # Keras predictions
    print("Getting Keras predictions...")
    keras_pred = keras_model.predict(X_test, verbose=0)
    keras_pred_classes = np.argmax(keras_pred, axis=1)
    
    # From scratch predictions
    print("Creating Simple RNN from scratch implementation...")
    rnn_scratch = SimpleRNNFromScratch(keras_model, vectorizer)
    
    print("Getting from scratch predictions...")
    scratch_pred = rnn_scratch.predict(X_test)
    scratch_pred_classes = np.argmax(scratch_pred, axis=1)
    
    # Calculate F1 scores
    keras_f1 = f1_score(y_test, keras_pred_classes, average='macro')
    scratch_f1 = f1_score(y_test, scratch_pred_classes, average='macro')
    
    print(f"\n=== COMPARISON RESULTS ===")
    print(f"Keras F1-Score: {keras_f1:.4f}")
    print(f"From Scratch F1-Score: {scratch_f1:.4f}")
    print(f"Difference: {abs(keras_f1 - scratch_f1):.6f}")
    
    # Detailed comparison
    print(f"\n=== DETAILED COMPARISON ===")
    print(f"Keras predictions shape: {keras_pred.shape}")
    print(f"From scratch predictions shape: {scratch_pred.shape}")
    
    # Show first 10 predictions
    print(f"\nFirst 10 predictions comparison:")
    print("Sample | Keras Class | Scratch Class | Match | True Label")
    print("-" * 55)
    for i in range(10):
        match = "✓" if keras_pred_classes[i] == scratch_pred_classes[i] else "✗"
        print(f"{i+1:6d} | {keras_pred_classes[i]:11d} | {scratch_pred_classes[i]:13d} | {match:5s} | {y_test[i]:10d}")
    
    # Calculate accuracy of implementation
    matches = np.sum(keras_pred_classes == scratch_pred_classes)
    accuracy = matches / len(keras_pred_classes)
    print(f"\nImplementation accuracy: {accuracy:.4f} ({matches}/{len(keras_pred_classes)} matches)")
    
    # Show probability differences for first 5 samples
    print(f"\n=== PREDICTION PROBABILITIES COMPARISON (First 5 samples) ===")
    for i in range(5):
        print(f"Sample {i+1}:")
        print(f"  Keras:       [{keras_pred[i][0]:.4f}, {keras_pred[i][1]:.4f}, {keras_pred[i][2]:.4f}]")
        print(f"  From Scratch: [{scratch_pred[i][0]:.4f}, {scratch_pred[i][1]:.4f}, {scratch_pred[i][2]:.4f}]")
        print(f"  Max diff:     {np.max(np.abs(keras_pred[i] - scratch_pred[i])):.6f}")
    
    return keras_f1, scratch_f1

# ====================================
# 5. MAIN EXECUTION (COMPLETED)
# ====================================

def main():
    """Main execution function"""
    print("="*60)
    print("SIMPLE RNN FOR TEXT CLASSIFICATION")
    print("="*60)
    
    # Load data
    train_df, test_df, valid_df = load_and_explore_data()
    
    # Preprocess data
    (X_train, y_train), (X_test, y_test), (X_valid, y_valid), vectorizer, label_encoder = preprocess_data(
        train_df, test_df, valid_df, max_features=5000, max_length=100
    )
    
    vocab_size = vectorizer.vocabulary_size()
    
    # Hyperparameter Analysis
    print("\n" + "="*60)
    print("MULAI ANALISIS HYPERPARAMETER")
    print("="*60)
    
    # 1. Analisis jumlah layer
    best_model_layers = analyze_num_layers(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    # 2. Analisis jumlah units  
    best_model_units = analyze_rnn_units(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    # 3. Analisis bidirectional
    best_model_direction = analyze_bidirectional(X_train, y_train, X_valid, y_valid, X_test, y_test, vocab_size)
    
    # ====================================
    # 6. FINAL MODEL TRAINING
    # ====================================
    
    print("\n" + "="*60)
    print("TRAINING FINAL MODEL DENGAN KONFIGURASI PALING BAGUS")
    print("="*60)
    
    final_model = create_rnn_model(
        vocab_size=vocab_size,
        embedding_dim=128,
        rnn_units=64,  # Berdasarkan analisis units
        num_rnn_layers=2,  # Berdasarkan analisis layers
        bidirectional=True,  # Berdasarkan analisis direction
        dropout_rate=0.3,
        num_classes=3,
        max_length=100
    )
    
    final_model, final_history, final_f1, final_predictions = train_and_evaluate_model(
        final_model, X_train, y_train, X_valid, y_valid, X_test, y_test,
        epochs=25, batch_size=32, model_name="Final_Best_RNN"
    )
    
    # ====================================
    # 7. CONFUSION MATRIX ANALYSIS
    # ====================================
    
    print("\n" + "="*60)
    print("ANALISIS CONFUSION MATRIX")
    print("="*60)
    
    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_test, final_predictions)
    
    plt.subplot(2, 2, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Negative', 'Neutral', 'Positive'],
                yticklabels=['Negative', 'Neutral', 'Positive'])
    plt.title('Confusion Matrix - Raw Counts')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # Normalized confusion matrix
    plt.subplot(2, 2, 2)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt='.3f', cmap='Blues',
                xticklabels=['Negative', 'Neutral', 'Positive'],
                yticklabels=['Negative', 'Neutral', 'Positive'])
    plt.title('Confusion Matrix - Normalized')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # Training history
    plt.subplot(2, 2, 3)
    plt.plot(final_history.history['loss'], label='Train Loss')
    plt.plot(final_history.history['val_loss'], label='Validation Loss')
    plt.title('Final Model - Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(2, 2, 4)
    plt.plot(final_history.history['accuracy'], label='Train Accuracy')
    plt.plot(final_history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Final Model - Training Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # ====================================
    # 8. FROM SCRATCH IMPLEMENTATION
    # ====================================
    
    print("\n" + "="*60)
    print("IMPLEMENTASI FROM SCRATCH")
    print("="*60)
    
    # Compare implementations
    keras_f1, scratch_f1 = compare_implementations(final_model, vectorizer, X_test, y_test)
    
    # ====================================
    # 9. FINAL SUMMARY AND INSIGHTS
    # ====================================
    
    print("\n" + "="*60)
    print("RINGKASAN HASIL DAN INSIGHT")
    print("="*60)
    
    print(f"\n=== PERFORMA MODEL FINAL ===")
    print(f"Final Model F1-Score: {final_f1:.4f}")
    print(f"Implementation Accuracy: {(keras_f1 - abs(keras_f1 - scratch_f1)):.4f}")
    
    # Class-wise performance analysis
    from sklearn.metrics import precision_recall_fscore_support
    precision, recall, f1_per_class, support = precision_recall_fscore_support(
        y_test, final_predictions, average=None
    )
    
    print(f"\n=== PERFORMA PER KELAS ===")
    class_names = ['Negative', 'Neutral', 'Positive']
    for i, class_name in enumerate(class_names):
        print(f"{class_name}:")
        print(f"  Precision: {precision[i]:.4f}")
        print(f"  Recall:    {recall[i]:.4f}")
        print(f"  F1-Score:  {f1_per_class[i]:.4f}")
        print(f"  Support:   {support[i]}")
    
    print(f"\n=== INSIGHT DAN ANALISIS ===")
    print("1. Hyperparameter Analysis:")
    print("   - Jumlah layer RNN optimal telah ditemukan melalui eksperimen")
    print("   - Jumlah units per layer telah dioptimalkan")
    print("   - Penggunaan Bidirectional RNN telah dievaluasi")
    
    print("\n2. Model Performance:")
    print(f"   - Macro F1-Score: {final_f1:.4f}")
    print("   - Model menunjukkan kemampuan klasifikasi yang baik")
    
    print("\n3. Implementation Verification:")
    print(f"   - Keras vs From Scratch difference: {abs(keras_f1 - scratch_f1):.6f}")
    print("   - Forward propagation implementation berhasil diverifikasi")
    
    # Find best and worst performing classes
    best_class_idx = np.argmax(f1_per_class)
    worst_class_idx = np.argmin(f1_per_class)
    
    print(f"\n4. Class Performance Analysis:")
    print(f"   - Best performing class: {class_names[best_class_idx]} (F1: {f1_per_class[best_class_idx]:.4f})")
    print(f"   - Worst performing class: {class_names[worst_class_idx]} (F1: {f1_per_class[worst_class_idx]:.4f})")
    
    # Recommendations
    print(f"\n=== REKOMENDASI UNTUK IMPROVEMENT ===")
    if f1_per_class[1] < 0.7:  # Neutral class often performs worst
        print("1. Neutral class performance dapat ditingkatkan dengan:")
        print("   - Data augmentation untuk class neutral")
        print("   - Teknik class balancing")
        print("   - Feature engineering yang lebih baik")
    
    if final_f1 < 0.8:
        print("2. Overall performance dapat ditingkatkan dengan:")
        print("   - Menggunakan pre-trained embeddings (Word2Vec, GloVe)")
        print("   - Mencoba arsitektur LSTM atau GRU")
        print("   - Ensemble methods")
        print("   - Hyperparameter tuning yang lebih ekstensif")
    
    print("3. Arsitektur alternatif yang bisa dicoba:")
    print("   - LSTM/GRU untuk mengatasi vanishing gradient")
    print("   - CNN-RNN hybrid architecture")
    print("   - Attention mechanisms")
    print("   - Transformer-based models")
    
    print(f"\n=== EKSPERIMEN SELESAI ===")

    
    return final_model, final_history, final_f1, vectorizer

# ====================================
# 10. ADDITIONAL UTILITY FUNCTIONS
# ====================================

def predict_new_text(model, vectorizer, label_encoder, text):
    """Fungsi untuk prediksi teks baru"""
    # Vectorize input text
    text_vectorized = vectorizer([text])
    
    # Predict
    prediction = model.predict(text_vectorized, verbose=0)
    predicted_class_idx = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class_idx]
    
    # Convert back to label
    idx_to_label = {v: k for k, v in label_encoder.items()}
    predicted_label = idx_to_label[predicted_class_idx]
    
    return predicted_label, confidence, prediction[0]

def interactive_prediction(model, vectorizer, label_encoder):
    """Fungsi interaktif untuk prediksi teks"""
    print("\n" + "="*50)
    print("INTERACTIVE PREDICTION")
    print("="*50)
    print("Masukkan teks untuk diprediksi (ketik 'quit' untuk keluar):")
    
    while True:
        text = input("\nMasukkan teks: ")
        if text.lower() == 'quit':
            break
            
        try:
            label, confidence, probabilities = predict_new_text(model, vectorizer, label_encoder, text)
            print(f"Prediksi: {label}")
            print(f"Confidence: {confidence:.4f}")
            print(f"Probabilities: Negative={probabilities[0]:.4f}, Neutral={probabilities[1]:.4f}, Positive={probabilities[2]:.4f}")
        except Exception as e:
            print(f"Error: {e}")

# ====================================
# 11. MAIN PROGRAM
# ====================================

if __name__ == "__main__":
    final_model, final_history, final_f1, vectorizer = main()
    
    label_encoder = {'negative': 0, 'neutral': 1, 'positive': 2}
    
    # Buat Coba Prediksi Manual
    # print("\nApakah Anda ingin mencoba prediksi interaktif? (y/n)")
    # choice = input().lower()
    # if choice == 'y':
    #     interactive_prediction(final_model, vectorizer, label_encoder)
```
